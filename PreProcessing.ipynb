```python
# Function to create a bar plot for target variable distribution
def plot_target_distribution(original_labels, y_train, y_test):
    labels_of_interest = [0, 1, 2, 3]

    fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

    for idx, y_data, title in zip(range(3), [original_labels, y_train, y_test], ['Original Labels', 'Train Dataset', 'Test Dataset']):
        unique_labels, label_counts = pd.Series(y_data).value_counts().sort_index().index, pd.Series(y_data).value_counts().sort_index().values
        bars = axes[idx].bar(unique_labels, label_counts, color='skyblue')
        axes[idx].set_title(f'{title} - Target Variable Distribution')
        axes[idx].set_xlabel('Label')
        axes[idx].set_ylabel('Count')
        axes[idx].set_xticks(labels_of_interest)

        # Add percentages on top of each bar
        for bar in bars:
            yval = bar.get_height()
            axes[idx].text(bar.get_x() + bar.get_width()/2, yval - 0.03 * max(label_counts), f'{yval / sum(label_counts) * 100:.2f}%', ha='center', va='bottom')

    plt.show()
```

### POS Tagging


```python
import re
DIGIT_RX = "($\d+([\.|,]\d+)?[\w]?[\s|-]){1,8}|[^A-Za-z\,()\.'\-:\!\? ]{1,8}"
SYMBOL_RX = "[/(/)\-/*/,\!\?]|[^ -~]"
DOT_RX = "\.{1,4}|\:"

train = pd.DataFrame(X_train, columns=['lyrics'])
train ['replaced_num']= train ['lyrics'].apply(lambda series: re.sub(DIGIT_RX, " zNUM ", series))
train ['replaced_symb'] = train ['lyrics'].apply(lambda series: re.sub(SYMBOL_RX, " ", series))
train ['replaced_dot'] = train ['lyrics'].apply(lambda series: re.sub(DOT_RX, " zDOT ", series))
train.tail()
```

```python
### Lemmatization & Removing Stop Words 

import spacy
from spacy.lang.en.stop_words import STOP_WORDS


nlp = spacy.load('en_core_web_sm')
nlp.Defaults.stop_words |= {" f ", " s ", " etc"}
stop_words = set([w.lower() for w in list(STOP_WORDS)])
```

```python
train['lemmatiz'] = train.replaced_dot.apply(lambda series: ' '.join([word.lemma_ for word in nlp(series)]))
train[['lyrics','lemmatiz']]
```

```python
import string

```

```python
train
```

```python
from tqdm import tqdm
# Define a function for lemmatization
def lemmatize_text(text):
    return ' '.join([word.lemma_ for word in nlp(text)])
# Apply lemmatization using tqdm for progress tracking
tqdm.pandas()
train['lemmatiz'] = train['replaced_dot'].progress_apply(lemmatize_text)
```

```python
#lemmatization +
tqdm.pandas()
train['contains_pron'] = train.lemmatiz.apply(lambda series: 1 if series.__contains__('-PRON-') else 0 )
```

```python
tqdm.pandas()
train['verb_count'] = train.lemmatiz.progress_apply(lambda series: len([token for token in nlp(series) if token.pos_ == 'VERB']))
```

```python
#remove punctuation
def remove_punct(series):
  table = str.maketrans('', '', string.punctuation)
  tokens_punct = series.translate(table).lower()
  tokens_spaces = ' '.join([token.strip() for token in tokens_punct.split() if token != ' '])
  return tokens_spaces
train['tokens_punct'] = train.lemmatiz.progress_apply(remove_punct)

```

```python
#remove stopwords
train['remove_stop_words'] = train.tokens_punct.progress_apply(lambda series: ' '.join([word for word in series.split() if word not in stop_words]))
```

```python
train['count_sent'] = train.remove_stop_words.apply(lambda series: 1 if series.count('zdot') == 0 else series.count('zdot'))
train['num_count'] = train.remove_stop_words.apply(lambda series: series.count('znum'))
train['count_words'] = train.remove_stop_words.apply(lambda series: len(nlp(series)))
```

```python
train = train[['lyrics', 'remove_stop_words', 'count_sent', 'num_count', 
               'count_words', 'verb_count','contains_pron']]
train
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(oov_token="<OOV>")
tokenizer.fit_on_texts(train['remove_stop_words'])

# Turn text into sequences (word --> num )
text_sequences = tokenizer.texts_to_sequences(train['remove_stop_words'])

# Turn text sequences into tf-idf sequences
tfidf_train = tokenizer.sequences_to_matrix(text_sequences , mode='tfidf')

print("The size of vocabulary",len(tokenizer.word_index)+1)
print("The nunber of rows in train set:",len(train))
print("The shape of tfidf_train sequences:",tfidf_train.shape)
```

### Pre-processing for Test Set

```python
test = pd.DataFrame(X_test, columns=['lyrics'])
```

```python
#POS-tagging for numbers and dots; replace symbols
test ['replaced_num']= test ['lyrics'].apply(lambda series: re.sub(DIGIT_RX, " zNUM ", series))
test ['replaced_symb'] = test ['lyrics'].apply(lambda series: re.sub(SYMBOL_RX, " ", series))
test ['replaced_dot'] = test ['lyrics'].apply(lambda series: re.sub(DOT_RX, " zDOT ", series))
```

```python
#lemmatization + count prons and verbs
test['lemmatiz'] = test.replaced_dot.progress_apply(lambda series: ' '.join([word.lemma_ for word in nlp(series)]))
test['contains_pron'] = test.lemmatiz.progress_apply(lambda series: 1 if series.__contains__('-PRON-') else 0 )
test['verb_count'] = test.lemmatiz.progress_apply(lambda series: len([token for token in nlp(series) if token.pos_ == 'VERB']))
```

```python
#remove punctuation
test['tokens_punct'] = test.lemmatiz.apply(remove_punct)

#remove stopwords
test['remove_stop_words'] = test.tokens_punct.apply(lambda series: 
                                                      ' '.join([word for word in series.split() if word not in stop_words]))
```

```python



#Linguistic statistics features
test['count_sent'] = test.remove_stop_words.apply(lambda series: 1 if series.count('zdot') == 0 else series.count('zdot'))
test['num_count'] = test.remove_stop_words.apply(lambda series: series.count('znum'))
test['count_words'] = test.remove_stop_words.apply(lambda series: len(nlp(series)))

test = test[['lyrics', 'remove_stop_words', 'count_sent', 'num_count', 
               'count_words', 'verb_count','contains_pron']]

# Turn text into sequences (word --> num )
text_sequences_test = tokenizer.texts_to_sequences(test['remove_stop_words'])

# Turn text sequences into tf-idf sequences
tfidf_test = tokenizer.sequences_to_matrix(text_sequences_test , mode='tfidf')

print("The size of vocabulary",len(tokenizer.word_index)+1)
print("The nunber of rows in train set:",len(test))
print("The shape of tfidf_train sequences:",tfidf_test.shape)
```
