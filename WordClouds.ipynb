```python
# Mapping labels to genres
label_to_genre = {0: 'Country',1: 'Indie', 2: 'Hip-Hop', 3: 'Rock'}

# Apply the mapping to the 'Label' column
df['Genre'] = df['Label'].map(label_to_genre)

# Display the updated DataFrame
df
```

```python
df.to_csv("Lyrics_Clean2.csv")
```

```python
from sklearn.model_selection import train_test_split

X = df['Lyrics'].copy()
y = df['Label'].astype(int).copy()
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    stratify=y, 
                                                    test_size=0.2, 
                                                    random_state=42)

print('The proportion in y_train\n', y_train.value_counts(normalize=True).mul(100).rename('label'))
print('The proportion in y_test\n', y_test.value_counts(normalize=True).mul(100).rename('label'))
print('The proportion in data\n', df.Label.value_counts(normalize=True).mul(100).rename('label'))
```

```python
X_train.to_csv("X_train.csv")
X_test.to_csv("X_test.csv")
y_train.to_csv("y_train.csv")
y_test.to_csv("y_test.csv")
```

```python
# Function to create and display word clouds for each genre
def create_word_cloud(genre, genre_lyrics):
    text = ' '.join(genre_lyrics)
    wordcloud = WordCloud(width=800, height=400, random_state=21, background_color='white').generate(text)

    plt.figure(figsize=(10, 7))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis('off')
    plt.title(f'Word Cloud for {genre}', fontsize=16)
    plt.show()

# Create word clouds for each genre
genres = df['Genre'].unique()
for genre in genres:
    genre_lyrics = df[df['Genre'] == genre]['remove_stop_words'].values
    create_word_cloud(genre, genre_lyrics)
```
